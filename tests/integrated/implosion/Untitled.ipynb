```python
r = np.linspace(1,2, 5)
theta = np.linspace(np.pi/2,0,5)

x = r*np.cos(theta)
y = r*np.sin(theta)
rho = np.ones((r.shape[0]-1, theta.shape[0]-1)) * np.nan
XX,YY = np.meshgrid(x,y)
plt.pcolormesh(XX,YY,rho,edgecolor="k",
    lw=0.1,
    antialiased=True)
plt.axis('equal')
```

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np


# Create the mesh in polar coordinates and compute corresponding Z.
r = np.linspace(1, 2, 10)
p = np.linspace(0, np.pi/2, 50)
R, P = np.meshgrid(r, p)
Z = np.zeros((R.shape[0]-1,R.shape[1]-1))
Z[0,:] = 1
Z[-1,:] = -1

Z[:,0] = 2
Z[:,-1] = -2

# Express the mesh in the cartesian system.
X, Y = R*np.cos(P), R*np.sin(P)

plt.pcolormesh(X,Y,Z,edgecolor="k",
    lw=0.1,
    antialiased=True)
plt.colorbar()
plt.axis('equal')
plt.show()
```
